### Let's use the pretrained VGG network 

```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = VGG16(weights='imagenet')
```

```python
img_path = '/content/cole-keister-cX-KEISwDIw-unsplash.jpg'
img_path1 = '/content/welsh-corgi.png'

img = image.load_img(img_path, target_size=(224, 224))
img1 = image.load_img(img_path1, target_size=(224, 224))

x = image.img_to_array(img)
x1 = image.img_to_array(img1)

x = np.expand_dims(x, axis=0)
x1 = np.expand_dims(x1, axis=0)

x = preprocess_input(x)
x1 = preprocess_input(x1)

preds = model.predict(x)
preds1 = model.predict(x1)
print(preds)
```

### compile the results into a list of tuples (class, description, probability)

```python
print('Predicted:', decode_predictions(preds, top=3)[0][0])
print('Predicted:', decode_predictions(preds1, top=3)[0][0])
```
