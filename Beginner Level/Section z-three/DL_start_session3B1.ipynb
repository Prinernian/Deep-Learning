### Load trained model

```python
from keras.models import load_model
model = load_model('cats_and_dogs_CNN_model.h5')
```

```python
from keras.preprocessing import image
from matplotlib.pyplot import imshow
import numpy as np
```

###your image path is a test image 

```python
img_path = '/content/test/dogs/dog1504.jpg'
```

```python
img = image.load_img(img_path, target_size=(150, 150))
imshow(img)
x = image.img_to_array(img)
x /= 255.
x = np.expand_dims(x, axis=0)
mp = model.predict(x)
print(mp)
print(np.round(mp).astype(int))
```
