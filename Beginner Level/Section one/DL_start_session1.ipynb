###convert an rgb image into grayscale function

````python
#Image processing technique
def rgb2gray(rgb):
  return np.dot(rgb[..., :3], [0.299, 0.587, 0.114]).astype(int)
````

````python
img = plt.imread("/content/cropped-3000-2000-1213559.jpg")
dataD = rgb2gray(img)
plt.imshow(dataD)
plt.show()
````

````python
dataset = io.loadmat('/content/Data_hoda_full.mat')
dataset
````

````python
X_train_original = np.squeeze(dataset['Data'][0:1000])
y_train = np.squeeze(dataset['labels'][0:1000])
X_test_original = np.squeeze(dataset['Data'][1000:1200])
y_test = np.squeeze(dataset['labels'][1000:1200])
````

````python
X_train_original
y_train
X_test_original
y_test
````

````python
i = 9
image = X_train_original[i]
label = y_train[i]
plt.imshow(image, cmap = 'gray')
plt.show()
print("label {}th is {}".format(i, label))
````

````python
#resize
X_train_5by5 = [resize(img, (5, 5)) for img in X_train_original]
X_test_5by5 = [resize(img, (5, 5)) for img in X_test_original]
````

````python
image = X_train_5by5[1]
plt.imshow(image, cmap = 'gray')
plt.show()
image.shape
````

````python
X_train = [x.reshape(25) for x in X_train_5by5]
X_test = [x.reshape(25) for x in X_test_5by5]
````

````python
X_test
````

````python
X_train[5].shape
````

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
X_train, X_test
y_train, y_test
````

````python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
````

````python
sample = 9
X = [X_test[sample]]
predicted_class = neigh.predict(X)
actual_classes = y_test[sample]
print("sample {} is a {}, and your prediction is :{}".format(sample, actual_classes, predicted_class))
````

````python
print(neigh.predict_proba(X))
````

````python
sample = 90
X = [X_test[sample]]
predicted_class = neigh.predict(X)
actual_classes = y_test[sample]
print("sample {} is a {}, and your prediction is :{}".format(sample, actual_classes, predicted_class))
````

````python
print(neigh.predict_proba(X))
````

````python
sample = 44
X = [X_test[sample]]
predicted_class = neigh.predict(X)
actual_classes = y_test[sample]
print("sample {} is a {}, and your prediction is :{}".format(sample, actual_classes, predicted_class))
````

````python
print(neigh.predict_proba(X))
````

````python
pred_classes = neigh.predict(X_test)
true_classes = y_test
print("predicted classes are :")
print(pred_classes)
print("true classes are :")
print(true_classes)
````

````python
#first way to see our accuracy score
from sklearn.metrics import accuracy_score
acc = accuracy_score(true_classes, pred_classes)
print(acc)
############################################################
#second way to evaluate our accuracy score with the KNN classifier
accuracy = neigh.score(X_test, y_test)
print(accuracy)
````
